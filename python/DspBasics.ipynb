## How to create a sine wave signal ?
There are various ways to generate a sine wave signal.

Because we are dealing with computer generated signals the signals are discrete in time and amplitude.

The values usually are output to loudspeaker with a sampling frequency 'fs' of 8000Hz :

```python
fs=8000 
```

We want to create a sine wave signal with the frequency 400Hz:

```python
f=400
```

Let's say we want to have 3 sine wave cycles:

```python
ncycles=3
numberOfPoints=fs/f*ncycles
```

To simplify the calculations we use the Matlab like Python toolbox called 'numpy':

```python
import numpy as np
nPointsVector=np.arange(numberOfPoints)
sinSig=np.sin(2*np.pi*f/fs*nPointsVector)
```

We can display the signal with the Python library 'matplotlib':

```python
import matplotlib.pyplot as plt

plt.plot(sinSig)
plt.title('sine wave signal')
plt.ylabel('amplitude')
plt.xlabel('n')
plt.grid()
plt.show()
```

An other way to create a sine wave signal is to create a time vector for a signal with a duration of 2 seconds:

```python
T = 2
#linspace arguments: startvalue,endvalue,number of points
t = np.linspace(0, T, T*fs, endpoint=False) 
sinSig=np.sin(2*np.pi*f*t)
```

We can play this sound with the Python library 'IPython':

```python
import IPython.display as ipd
ipd.Audio(sinSig, rate=fs) 
```

Now we want to use an Fast Fourier Transform ( FFT ) to analyse the frequency content of the signal:

```python
sp = np.fft.fft(sinSig)

realPartOfSignal=sp.real
absoluteSignal=np.absolute(sp)

plt.plot(absoluteSignal)
plt.title('fft absolute value')
plt.ylabel('amplitude')
plt.xlabel('n')
plt.grid()
plt.show()
```
